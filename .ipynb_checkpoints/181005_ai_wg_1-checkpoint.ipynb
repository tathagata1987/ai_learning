### Deep Neural Network Architecture

The problem may be formulated with 7 input neurons representing the rating of each input data points, and an output layer with two sets of 7 softmax neurons representing respectivelly the probabilities and the subordinations of a credit portfolio loss distribution. Remember that each of these 7 softmax neurons yield a probability (number between 0 and 1) and the 7 probabilities add up to 1. Once trained, this network would result in a discrete loss probability distribution. Then a credit spread may be easily computed for each rating cohort from the the loss probabilities of each rating.



Source: Drawn from http://alexlenail.me/NN-SVG/index.html

The trick is on formulating the problem with both input and output data as well as the cost function. In the context of the network above, the input data would be an array of zeros and ones with 7 positions, with 1 representing the rating of the specific data points. The output labels would be mostly zeros and one position with the default probability of the respective rating data point from the input.
